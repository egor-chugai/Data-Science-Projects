## 2. Обучение и проверка модели для каждого региона. <a name="2."></a>  
[к содержанию](#0.0)

* Разобьем данные на обучающую и валидационную выборки в соотношении 75:25.
* Обучим модель и сделаем предсказания на валидационной выборке.
* Посчитаем RMSE моделей.

```python
# в цикле обучим модель на каждом датасете и посчитаем RMSE
state=np.random.RandomState(123)
results = []
splitted_datasets = []

for dataset in range(len(datasets)):
    
    # разделим выборку
    features = datasets[dataset].drop(['id', 'product'], axis=1)
    target = datasets[dataset]['product']
    
    features_train, features_valid, target_train, target_valid =\
        train_test_split(features, target, test_size=0.25, random_state=state)

    # обучим модель линейной регрессии, сделаем предсказание
    model = LinearRegression()
    model.fit(features_train, target_train)
    prediction = model.predict(features_valid)
    
    # посчитаем RMSE и средние значения
    MSE = mean_squared_error(target_valid, prediction)
    RMSE = sqrt(MSE)
    predicted_mean = prediction.mean()
    actual_mean = target.mean()
    
    # добавим результаты в словарь
    results.append({
        'Region_name' : datasets_names[dataset],
        'RMSE' : RMSE,
        'predicted_mean' : predicted_mean,
        'actual_mean' : actual_mean})

    # добавим разделенные датасеты в словать, для переиспользования в дальнейшем
    splitted_datasets.append({
        'Region_name' : datasets_names[dataset],
        'features_train': features_train, 
        'features_valid': features_valid, 
        'target_train': target_train, 
        'target_valid': target_valid,
        'prediction': pd.Series(prediction, index=target_valid.index)})

```

```python
pd.DataFrame(results)
```

### Вывод

После построения моделей отметим следующее:
* Наилучший показатель RMSE наблюдается во втором регионе (регион с наименьшим запасом нефти и большим к-вом пустых скважин);
* Среднее на предсказанных значениях мало отличается от реальных значений.


## 3. Подготовка к расчету прибыли. <a name="3."></a>  
[к содержанию](#0.0)

Все ключевые значения для расчётов сохраним в отдельных переменных

```python
All_points = 500 #кол-во точек
Best_points = 200 #кол-во луших точек
Region_budget = 10000000000 #бюджет на разработку скважин в регионе
Profit_per_barrel = 450000 #доход с каждой ед. продукта в рублях
Risk_treshold = 0.025 #максимальный трешхолд убытков
```

Рассчитаем достаточный объём сырья для безубыточной разработки новой скважины.
Сравним полученный объём сырья со средним запасом в каждом регионе.

```python
budget_new_point = Region_budget / Best_points # бюджет на одну скважину
print('Бюджет на одну скважину:', budget_new_point)
```

```python
print('Объём сырья для безубыточной разработки новой скважины:', budget_new_point / Profit_per_barrel)
```

Чтобы не нести убытки скважина должна распологать объемом не менее 111.1 баррелей.

### Вывод

Полученный объём сырья для безубыточной разработки новой скважины - 111.1, существенно выше средних значений во всех трех регионах.

## 4. Функция для расчёта прибыли <a name="4."></a>  
[к содержанию](#0.0)

Напишем функцию для расчёта прибыли по выбранным скважинам и предсказаниям модели. Функция будет принимать 3 аргумента:
* target - истинные значения целевого признака;
* predictions - предсказанное значение целевого признака;
* number_of_wells - количество скважен.

```python
def revenue(target, predictions, number_of_wells):
    
    # выберем индексы с максимальными значениями предсказаний
    max_predicted_values = predictions.sort_values(ascending = False).head(number_of_wells).index
    
    # просуммируйте целевые значение объёма сырья, соответствующее этим предсказаниям
    # посчитаем суммарный доход с этих скважен
    income = target[max_predicted_values].sum() * Profit_per_barrel
    
    # расчитаем прибыль
    revenue_final = income - Region_budget
    return revenue_final
```

## 5. Расчет рисков и прибыли для каждого региона <a name="5."></a>  
[к содержанию](#0.0)

Для анализа и просчета рисков выполним следующие шаги:
* применим Bootstrap для нахождения распеределения прибыли,
* найдем среднюю прибыль, 95%-й доверительный интервал и риск убытков,
* напишем вывод и предложим регион для разработки скважин

Для расчета напишем цикл, для удобства переведем полученные результаты в миллионы рублей.

```python
bootstrap_res = []
revenue_region_values = []


for i in range(len(splitted_datasets)):
    target_valid = splitted_datasets[i].get('target_valid')
    prediction = pd.Series(splitted_datasets[i].get('prediction'), index=target_valid.index)

    # применим технику Bootstrap с 1000 выборок, для нахождения распределения прибыли
    revenue_values = []
    for j in range(1000):
        target_subsample = target_valid.sample(n=All_points, replace=True, random_state=state)
        predict_subsample = prediction[target_subsample.index]
        revenue_values.append(revenue(target_subsample, predict_subsample, Best_points))
    revenue_values = pd.Series(revenue_values) 

    # посчитаем среднюю прибыль и построим доверительные интервалы
    mean = revenue_values.mean() / 1000000 # средняя прибыль
    lower = round(revenue_values.quantile(Risk_treshold)  / 1000000, 1) # начало доверительного интервала
    upper = round(revenue_values.quantile(0.975) / 1000000, 1) # конец доверительного интервала
    loss_risks = (revenue_values < 0).sum() / revenue_values.count() # риск убытков
    
    # занесем результаты анализа в словари
    bootstrap_res.append({
        'Region': datasets_names[i], 
        'lower': lower, 
        'upper': upper,
        'loss_risks': loss_risks,
        'mean': mean
    })
    
    # занесем распределения прибыли в словари
    revenue_region_values.append({
        'Region': datasets_names[i],
        'values': revenue_values
    })
```

Выведем гистограммы распределения прибыли и результаты в таблице

```python
plt.figure(figsize=(18, 5))
p = 131

for i in range(3):
    plt.subplot(p)
    sns.distplot(revenue_region_values[i].get('values'))
    plt.title(datasets_names[i])
    p += 1
plt.show()

bootstrap_results = pd.DataFrame(bootstrap_res)
bootstrap_results
```

Наиболее рискованным регионом для разработки оказался третий регион, в нем риски убытков **4,5%** вместо допустимых 2,5%. Наименее рискованный - Второй регион 0,1% риска убытков. Также во стором регионе наивысшее показатель средней прибыли - 693 млн.рублей. Таким образом, для разработки можно предложить второй регион.
